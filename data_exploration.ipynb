```python
import os
import csv
import time
import h5py
import cv2
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
SRC_PATH = '/home/ashish/Dev/CarND-Behavioral-Cloning-P3'
DATA_PATH = '/home/ashish/Dev/CarND-Behavioral-Cloning-P3-data'
RUNS = ['run11','run12','run14','run30','run31','run16','run1','run2','run3']
```

```python
lines = {}

for idx, run in enumerate(RUNS):
    run_path = os.path.join(DATA_PATH, run)
    driving_log_path = os.path.join(run_path, 'driving_log.csv')
    temp = []
    with open(driving_log_path, 'r') as csvfile:
        reader = csv.reader(csvfile)
        for line in reader:
            temp.append(line)
    lines[run] = temp

lines.keys()
```

```python
steering_angles = {}
for run, samples in lines.items():
    temp = []
    for sample in samples:
        temp.append(float(sample[3].strip()))
    steering_angles[run] = temp
```

```python
for run, angles in steering_angles.items():
    print(run)
    plt.figure(figsize=(10,10))
    sns.distplot(angles, kde=False)
plt.show()
```

```python
# Percent non zero steering angles
for run, angles in steering_angles.items():
    angles_np = np.array(angles)
    print(run, len(angles_np[angles_np != 0.0])/len(angles_np))
```
